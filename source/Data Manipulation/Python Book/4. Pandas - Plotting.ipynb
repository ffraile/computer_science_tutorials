### How to plot a DataFrame using Pandas 
Data Visualization is an important utility for every engineer, especially when you have to deal with huge datasets. Data vizualisation is a perfect tool to use to understand your dataset,to capture trends, to start analyzing it and of course to present to the others your conclusions. Pandas provides some options for plotting that are easy and straightforward to code. 

#### Some Plot Types in Pandas:
- line
- bar
- scatter
- histogram
- boxplot
- pie plot and so on

You can just use one line of code 
```df.plot(kind = 'type_name') ``` and replace the kind argument with the kind of plot you want to create.



Let's use a data set which contains data for admission in university and the chance of admission. We need to first import pandas library.
We would advise to also add : ``` %matplotlib inline ```
This command is a magic command which ensures the plotted figures appear in the notebook cells in a correct way.

````python
#import libraries
import pandas as pd
%matplotlib inline

#import dataset from csv file
df = pd.read_csv("C:/Users/belin/Documents/python/Data/adm_data.csv")

#take a little insight of how our dataframe looks like
df.head()
````

### Line Plot:
- used to track changes over a period of time     

The default line plot take as: 
- numeric columns into y-axis 
- index on the x-axis
When you want to plot only one specific column, in our case GRE Scores, you need to specify it.
``` (y="name_of_col")```


```figsize(width_size, height_size)``` 
- Sets the size of the plot( width and height in inches respectively)


````python
# to plot a line plot
df.plot(y="GRE Score", figsize= (8,4))
````

If you want to add:
- a title you need to add:  ``` title= "title_name" ```
- ylabel name : ``` ylabel= "ylabel_name" ```

#### Task
Add the line plot of GRE Scores a title and label the y-axis.

````python
#Add arguments such as title, labels, legend
df.plot(y="GRE Score", figsize= (8,4),title = "GRE Scores", ylabel = "GRE")
````

In order to plot more than one column, you can just specify the column names and assign it to the y-axis
```python
df.plot.line(y=['name_of_col1','name_of_col2'])
```

````python
df.plot.line(y=['SOP','CGPA'])
````

### BAR PLOT 
- used to show comparisons among discrete categories
 
A bar plot is used to present categorical data with rectangular bars with heights or lengths proportional to the values that they represent.



Our dataset contains Fertility Rate for different countries throughtout the years.
Let's import first our Excel File and then start plotting.

````python
df2 = pd.read_excel("C:/Users/belin/Documents/python/Data/Fertility-Rate.xls")
df2
````

#### Task
Plot a whole dataframe to a bar plot. 

y-axis shows the Fertility Rate and x-axis shows the countries but by index because it automatically gets the first column of the dataset as our x-axis.
Each year is represented into different colours as shown in the plot below.

````python
df2.plot.bar()
````

If you want to plot a specific column, you need to specify it as shown below. 
```python 
df_name.plot.bar(x="x-axis_col", y = "y-axis_col", color = "color_name")
```

````python
df2.plot.bar(x="Country Name", y = "1999", color = "red")
````

If you want to create separate subplot, you can add an argument ```subplots=True``` to the function. 


````python
df2.plot.bar(x="Country Name", y = ["1999","2005","2011"], subplots=True)
````

### Pie Plot

Helps in organizing and showing data as a percentage of a whole. 
Not advised to use if you want to show changes over time (unlike Bar Graphs or Line Graphs)

#### TASK
Show which country had the most Fertility rate in year 2011.
```python
df_name.plot(labels =df_name["x_col"], y = "y_col", kind='pie', autopct='font_size', title='title_name',legend="True or False depending on your preferences)
```

(Hint Use autocpt to display percentages in pie chart)

````python
df2.plot(labels =df2["Country Name"], y = "2011", kind='pie', autopct='%1.1f%%', title='Fertility Rate in 2011',legend=False)
````

### Scatter Plot
Show the relationship between 2 numerical variables.


#### Task 
Create a Scatter Plot to visualize the Fertility Rate in 2003.
The colors of the points should be red and the x-axis should be readable.


- c-> specify the color of the points
- rot -> specify the degree you want to rotate the labels of x-axis

````python
df2.plot(x = "Country Name", y = "2003",kind='scatter', c = 'red', rot  = 30)
````

### Histogram 

- Visualize the distribution of numerical data
- divides the values within a numerical variable into "bins" and count the number of observations that fall into each bin. 
- Shows the number of occurrences of different values in a dataset

```python
df.plot(kind='hist', y= "col_name")
```

##### Task
Plot a Histogram that shows the distribution of GRE Scores

````python
df.plot(kind='hist', y= "GRE Score")
````

##### Bins -> groups of values that fall into the same range.
- Default number is 10 but you can change and specify according to the range you want. 

````python
 df.plot(kind='hist', y= "GRE Score", bins=30, title='Scores GRE',figsize=(10,7),grid=True)
````

### Boxplot
- is used to show statistical data based on Minimum, first Quartile, Median, Third Quartile, and Maximum.
It is formed by a box from Q1 to Q3 quartile values of the data and the median(Q2) is represented by a line.
The boxplot shows outliers and what their values are. 
You can understand if the data is symmetrical, how is it skewed and also how tightly the data is grouped. 

````python
df.boxplot(column="TOEFL Score")
````
