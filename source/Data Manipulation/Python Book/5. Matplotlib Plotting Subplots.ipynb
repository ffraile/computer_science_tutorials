### Create Multiple Subplots with Matplotlib

Sometimes you want to show different columns each of them represented in a separate plot. Subplots are your way to go!

```python
pyplot.subplots 
```
can be used to create a figure which contains a grid of subplots.


#### Task
 Take the first 100 obersevations of GRE, TOEFL, Serial Number, SOP, University Ranking and assign each of them into a variable

````python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

#import dataset from csv file
df = pd.read_csv("C:/Users/belin/Documents/python/Data/adm_data.csv")

#take only the first 100 observation of each interested column 
GRE = df['GRE Score'][:100]
TOEFL = df['TOEFL Score'][:100]
x = df["Serial No."][:100]
SOP = df['SOP'][:100]
UNI_R = df['University Rating'][:100]

#see how your dataset looks like
df.head()
````

```
fig, ax = plt.subplots()
```
- return a Figure and a single Axes that we will populate later.
You should think of the fig as the whole image and ax as parts of the fig.

````python
fig, ax = plt.subplots()
#Set the title of the plot as "Empty Plot"
ax.set_title('Empty Plot')
````

### Creating 2 subplots 

#### Task
Plot 2 Line Plots showing GRE Scores and TOEFL scores of the students.

###### Steps:
specify the number of subplots by: 
- fig, axs = plt.subplots(nr_of_subplots)
Set the Title for the whole figure
- fig.suptitle("Title here")

Think the axs[] as a vector and you can move through different subplots by using the index of the vector
```python
 axs[nr_of_plot].plot(x_column, y_column)
```


````python
fig, axs = plt.subplots(2)
#set the title
fig.suptitle('Student Scores Overview')
#specify the first subplot
axs[0].plot(x,TOEFL)
#specify the second subplot
axs[1].plot(x,GRE)
````

### Horizontal Stacked Subplots

In order to plot Horizontal Stacked Subplots, you need to 
- pass parameters (row_nr,column_nr) to plt.subplots() 

In our case where we have only 2 subplots,it means:(1,2) for 1 row, 2 columns

 {axs[1] is the same as ax1}
 
 You can specify the size of the plots by adding 
 - figsize=(h,w)

````python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,5))
fig.suptitle('Student Scores Overview')
ax1.plot(x,TOEFL)
ax1.set_title("TOEFL Scores")
ax2.plot(x,GRE)
ax2.set_title("GRE Scores")
````

#### Task
Create 1 plot "Student Scores Overview" which contains 4 subplots for 

````python
fig, axs = plt.subplots(2, 2)
#Title of the Complete Graph
fig.suptitle('Student Scores Overview')

#First plot
axs[0, 0].plot(x, TOEFL)
axs[0, 0].set_title('TOEFL')

#Second Plot
axs[0, 1].plot(x, GRE, 'tab:red')
axs[0, 1].set_title('GRE')

#Third Plot
axs[1, 0].plot(x, SOP, 'tab:pink')
axs[1, 0].set_title('SOP')

#Forth Plot
axs[1, 1].plot(x, UNI_R, 'tab:green')
axs[1, 1].set_title('University Ratings')
````

#### Add Space Between Subplots
If you want to add Space between subplots, you can use 
- tight_layout()
By Default it will maintain the proper space between subplots but if you want to be specific then you can specify it by writing
- tight_layout(pad=nr_of_padding)

````python
fig, axs = plt.subplots(2, 2)
#Title of the Complete Graph
fig.suptitle('Student Scores Overview')
#Set the Space between subplots
fig.tight_layout()

#First plot
axs[0, 0].plot(x, TOEFL)
axs[0, 0].set_title('TOEFL')

#Second Plot
axs[0, 1].plot(x, GRE, 'tab:red')
axs[0, 1].set_title('GRE')

#Third Plot
axs[1, 0].plot(x, SOP, 'tab:pink')
axs[1, 0].set_title('SOP')

#Forth Plot
axs[1, 1].plot(x, UNI_R, 'tab:green')
axs[1, 1].set_title('University Ratings')
````

Sets the spacing between subplots 
- subplots_adjust() 


````python
fig, axs = plt.subplots(2, 2)
#Title of the Complete Graph
#Set the Space between subplots
fig.tight_layout()

#First plot
axs[0, 0].plot(x, TOEFL)
axs[0, 0].set_title('TOEFL')

#Second Plot
axs[0, 1].plot(x, GRE, 'tab:red')
axs[0, 1].set_title('GRE')

#Third Plot
axs[1, 0].plot(x, SOP, 'tab:pink')
axs[1, 0].set_title('SOP')

#Forth Plot
axs[1, 1].plot(x, UNI_R, 'tab:green')
axs[1, 1].set_title('University Ratings')

# set the spacing between subplots
plt.subplots_adjust(left=0.2,bottom=0.2,right=1,top=1,wspace=0.4,hspace=0.4)
````

### Sharing Axes

Sometimes when you create subplots, they may share one of the axes.
In this case you do not have to write the values of x or y twice. 
It is enough for only one graph to have them. 

In our dataset, we can see that SOP and University Rankings have the same range of values for y-axis.
We need to add 
- sharey=True 
at subplots() for the vertical axis align 

(use sharex=True if you want to align the horizontal axis)


##### Task
- Create 2 subplots that share the same y-axis and show the tendency of SOP and University Rankings
- Set a title for the whole plot and also a title for each subplot
- Change the color of one of the subplots

````python
#create 2 subplots that share the same yaxis
fig, (ax1, ax2)= plt.subplots(1, 2, sharey=True)
#set the title for the plot
fig.suptitle('SOP AND UNIVERSITY RANKINGS OF 100 Students')
#create the 1st subplot and set the title
ax1.plot(x, SOP)
ax1.set_title("SOP")
#create the 2nd subplot and set the title
ax2.plot(x, UNI_R, 'tab:red')
ax2.set_title("Univeristy Ranking")

````
