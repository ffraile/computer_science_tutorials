### Exercise 2
Review the following code:

```python
text = input("Enter some text: ")

text_2 = text[::-1]
if text == text_2:
    print("True!")
else:
    print("No luck this time...")
```
**a.** Describe in your words what the program does? **(1.5 points)**
This program takes a string from the user and checks if it is a palindrome. A palindrome is a word, phrase, or sequence that reads the same backward as forward, considering commas, spaces, and capital letters. For instance, the input text "hello" will result in the program printing "No luck this time..." while the input text "racecar" will result in the program printing "True!".

**b.** What is the output of the program if the user enters the text ```A man, a plan, a canal, Panama```? Explain why **(0.5 points)**
The output of the program is "No luck this time...". Although the text is a palindrome, the program does not consider commas and spaces, it just compares the text to the reverse of the text, and "A man, a plan, a canal, Panama" is not the same as "amanaP ,lanac a ,nalp a ,nam A", although reading the latter backwards results in the former.

### Exercise 4
Check the following program:

```python
numbers = [2, 7, 4, 9, 12, 5, 15]
for num in sorted(numbers, reverse=True):
    if num % 2 == 0:
        print(num)
```

Recall that the function ```sorted``` returns a sorted list of the elements in the list passed as argument. The argument ```reverse=True``` indicates that the list should be sorted in descending order (from largest to smallest).

**a.** What does the program do? Explain in your own words **(1 point)**
This program prints all the even numbers in the list numbers in descending order.

**b.** What is the sequence printed by the program? **(0.5 points)**
The sequence printed by the program is 12, 4, 2 (one number per line).
