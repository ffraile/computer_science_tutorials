### Exercise 3
Review the following program code:

```python
import random
all_chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!?.:,'{}[]+-*/=_@#$%^&()~"

p_len = int(input("Enter the length: "))
p_string = ""
for i in range(p_len):
    a_random_idx = random.randint(0, len(all_chars)-1)
    p_string = p_string + all_chars[a_random_idx]

print(p_string)
```

And recall that the function ```random.randint(a, b)``` returns a random integer between `a` and `b` (inclusive), and the function ```range(a)``` returns an iterable object that contains all integers from 0 to `a-1` (inclusive).

Answer the following questions:

**a.** Describe in your own words what the code does. What would you use this code for? **(2 points)**

### Exercise 5
Review the following program code:

```python
x = int(input("Enter an integer number: "))
i = 0
result = ""
while x > 2**i:
    b = (x // (2**i)) % 2
    i = i + 1
    result = str(b) + result
print(result)
```
