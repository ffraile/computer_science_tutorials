## Exercise 1 (2.5 points)
Review the following code:

```python
favorite_color = "blue"
color_guess = ""
attempt_limit = 3
attempts = 0
clues = ["It's a color of the sky.", "Superman wears this color.", "It's a primary color."]
while (color_guess.lower() != favorite_color) and (attempts < attempt_limit):
    color_guess = input("Guess my favorite color: ")
    attempts += 1
    if color_guess.lower() == favorite_color:
        print("Correct! That's my favorite color.")
    else:
        print("Sorry, try again.")
        print ("Hint: " + str(clues[attempts - 1]))
        attempts_left = attempt_limit - attempts
        print("Attempts left: " + str(attempts_left))

if attempts == attempt_limit:
    print("Sorry, you've run out of attempts.")
```

Note that the method lower() is used to convert the user's input to lowercase (same characters but no capital letters).

- **b.** What is the output of the program if the user's guesses are Red, green, and Blue in that order? (0.5 points)

## Exercise 2 (2.5 points)
Review the following code:

```python
n = int(input("Enter the number of terms in the harmonic series: "))
total_sum = 0
for i in range(1, n + 1):
    total_sum += i / (i + 1)

print(f"The sum of the series is {total_sum:.2f}")
```
