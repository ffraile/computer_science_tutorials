## Problem definition
The Thue Morse Sequence is a binary sequence that is generated by negating (binary complement) the bits in the sequence so far, and appending the resulting bits to the sequence, starting with 0. That is, the sequence is built iteratively like this:
0,
01,
0110,
01101001,
...
Notice that the new bits added in every sequence step are the binary complement of the sequence. You can find more information about the Thue-Morese Sequence [here](https://en.wikipedia.org/wiki/Thue%E2%80%93Morse_sequence).
Can you create a Thue-More sequence generator in Python?

## Solution
Our strategy is based on performing bitwise operations with integer variables to calculate the binary complement, then converting to strings to easily append the result to the sequence:

```python
n_steps = int(input("Specify the number of steps: "))
thue_morse = "0"
for i in range(n_steps):                                    # Repeat n_steps times
    binary_complement = ""                                  # Initialise an empty string
    for j in range(len(thue_morse)):                        # Then, for every bit in the sequence so far
        binary_complement += str(int(thue_morse[j]) ^ 1)    # Calculate the binary complement using bitwise xor,
                                                            # and  convert the result to a string
    thue_morse += binary_complement

print(thue_morse)
```

In this solution, we calculate the binary complement bit by bit to ensure that the leading zero is also taken into account.
